```python
import os
import glob
import pandas as pd
```

```python
file_extension = '.csv'
cw = os.path.abspath("Transcripts/completed_words/") + "/"
cs = os.path.abspath("Transcripts/completed_sentences/") + "/"
ct = os.path.abspath("Transcripts/completed_transcripts/") + "/"
wfilepath = os.path.abspath("Transcripts/word_transcripts/") + "/"
sfilepath = os.path.abspath("Transcripts/sentence_transcripts/") + "/"
tfilepath = os.path.abspath("Transcripts/full_transcripts/") + "/"

def find_csv_filenames(path_to_dir, suffix=".csv"):
    filenames = os.listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith(suffix)]

def prepend_path(filenames, directory):
    new_list = []
    for item in filenames:
        new_filename = directory + item
        new_list.append(new_filename)
    return new_list
```

```python
w1 = find_csv_filenames(wfilepath)
s1 = find_csv_filenames(sfilepath)
t1 = find_csv_filenames(tfilepath)

words = prepend_path(w1, wfilepath)
sentences = prepend_path(s1, sfilepath)
transcripts = prepend_path(t1, tfilepath)
```

```python
merged_words = pd.concat([pd.read_csv(f, usecols=['word','confidence','start_time','end_time','sentence','sentence_confidence','date','time','original_audio']) for f in words])
merged_words.to_csv(cw + 'merged_words.csv', index=False)

merged_sentences = pd.concat([pd.read_csv(f, usecols=['sentence','word_count','confidence','date','time','original_audio']) for f in sentences])
merged_sentences.to_csv(cs + 'merged_sentences.csv', index=False)

merged_transcripts = pd.concat([pd.read_csv(f, usecols=['transcript','word_count','date','time','original_audio']) for f in transcripts])
merged_transcripts.to_csv(ct + 'merged_transcripts.csv', index=False)
```

### Add 'merged_' to 'completed_'

```python
w1 = find_csv_filenames(cw)
s1 = find_csv_filenames(cs)
t1 = find_csv_filenames(ct)

words2 = prepend_path(w1, cw)
sentences2 = prepend_path(s1, cs)
transcripts2 = prepend_path(t1, ct)
```

```python
full_words = pd.concat([pd.read_csv(f) for f in words2])
full_words.to_csv(cw + 'words.csv', index=False)

full_sentences = pd.concat([pd.read_csv(f) for f in sentences2])
full_sentences.to_csv(cs + 'sentences.csv', index=False)

full_transcripts = pd.concat([pd.read_csv(f) for f in transcripts2])
full_transcripts.to_csv(ct + 'transcripts.csv', index=False)
```
